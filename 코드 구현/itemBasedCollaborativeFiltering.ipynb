## Data Preprocessing (데이터 전처리)

```python
rating_data.drop('timestamp', axis = 1, inplace=True)    # inplace로 update
rating_data.head(2)
```

```python
user_movie_rating = pd.merge(rating_data, movie_data, on = 'movieId')    # 공통 columnd인 'movieId'를 기준으로 merge
user_movie_rating.head(2)
```

##Item based collaborative filtering 위해 pivot table 생성

```python
movie_user_rating = user_movie_rating.pivot_table('rating', index = 'title', columns = 'userId')    # title이 index인 피벗테이블
user_movie_rating = user_movie_rating.pivot_table('rating', index = 'userId', columns = 'title')    # userId가 index인 피벗테이블
```

```python
user_movie_rating.head()
```

```python
movie_user_rating.head()
```

```python
# 아이템 기반 협업 필터링이기 때문에 movie(item) title을 index로 설정
# NaN 값을 0으로 채워줌
movie_user_rating.fillna(0, inplace = True)
movie_user_rating.head(3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_based_collabo = cosine_similarity(movie_user_rating)    # 코사인 유사도 계산
item_based_collabo
```

```python
item_based_collabo = pd.DataFrame(data = item_based_collabo, index = movie_user_rating.index, columns = movie_user_rating.index)
```

```python
item_based_collabo.head()
```

## table 기반으로 top-k 추천 함수 생성

```python
def top_k_items(title, k):
  return item_based_collabo[title].sort_values(ascending=False)[:(k+1)]
```
